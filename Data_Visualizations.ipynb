```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

### LOAD datasets

MP_age = pd.read_csv("MPs_by_Age.csv", index_col=0)
data = pd.read_csv("Voters_and_MPs_Age_buckets_per_Parliament_June14-2021.csv", index_col=0)
```

### VISUALIZATIONS

```python
plt.figure(figsize=(20, 8))
hue_order = ["18-30","31-40","41-50","51-60","61-70","71+"]
sns.barplot(x="Parliament", 
            y="Age@Parliament", 
            hue="Age_buckets", 
            data=MP_age,
            hue_order=hue_order)
plt.ylabel("Age@Parliament", size=14)
plt.xlabel("Parliament", size=14)
plt.title("Age buckets per Parliament", size=18)
```

```python
plt.figure(figsize=(20, 8))
hue_order = ["18-30","31-40","41-50","51-60","61-70","71+"]
sns.lineplot(x="Parliament_Year", 
            y="Age@Parliament", 
            hue="Age_buckets",
            style="Age_buckets",
            data=MP_age,
            markers=True, dashes=False)
#             hue_order=hue_order, markers=True)
plt.ylabel("Age@Parliament", size=14)
plt.xlabel("Parliament", size=14)
plt.title("Age buckets per Parliament", size=18)
```

```python
plt.figure(figsize=(30, 20))
sns.factorplot(x='Parliament_year', y='Voters_count', hue='Age_buckets', data=data)
sns.factorplot(x='Parliament_year', y='MP_count', hue='Age_buckets', data=data)
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x="Parliament_year", hue="Age_buckets", y="Voters_count", data=data)
plt.title("Voters by Age buckets per Parliament", size=18)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.barplot(data= data,
                x="Parliament_year",
                y="MP_count_%",
                hue="Age_buckets",
                )
plt.title("MPs by Age buckets per Parliament", size=18)
```

```python
## PLOTS WITH SHARED Y AXIS
# specify plot layouts with different width using subplots()
f, axs = plt.subplots(1,2,
                      figsize=(16,10),
                      sharex=True)
# make scatterplot with legends
sns.barplot(data= data,
                x="Parliament_year",
                y="MP_count",
                hue="Age_buckets",
                ax=axs[0],
                )
# make densityplot with kdeplot without legends
sns.barplot(data= data,
                x="Parliament_year",
                y="Voters_count",
                hue="Age_buckets",
                ax=axs[1])
f.tight_layout()
```

```python
## PLOTS WITH SHARED Y AXIS
# specify plot layouts with different width using subplots()
for k in data["Parliament_year"].unique():
    
    f, axs = plt.subplots(1,2,
                          figsize=(12,8),
                          sharey=True)
    # make scatterplot with legends
    sns.barplot(data= data[data["Parliament_year"]==k],
                    x="Parliament_year",
                    y="MP_count_%",
                    hue="Age_buckets",
                    ax=axs[0],
                    )
    # make densityplot with kdeplot without legends
    sns.barplot(data= data[data["Parliament_year"]==k],
                    x="Parliament_year",
                    y="Voters_count_%",
                    hue="Age_buckets",
                    ax=axs[1])
    f.tight_layout()
    plt.title("Age buckets (%) at {} Parliament".format(k), size=18)
```

```python
plt.figure(figsize=(20, 12))
sns.scatterplot(data= data, x="MP_count_%", y="Voters_count_%", hue="Age_buckets");
plt.savefig("MP_vs_Voters_per_Age_bucket_new")
```

```python
cond_plot = sns.FacetGrid(data=data, col='Parliament_year', hue='Age_buckets', col_wrap=4)
cond_plot.map(sns.scatterplot, 'Voters_count_%', 'MP_count_%');
```

```python
plt.figure(figsize=(18,10))
sns.stripplot(data = data, x = 'Age_buckets', y = 'MP_count_%')
sns.stripplot(data = data, x = 'Age_buckets', y = 'Voters_count_%', edgecolor="black",
                linewidth=2,sizes=(40,100))
```

```python
## SWARM PLOT - Along with the number of data points, it also provides their respective distribution.
plt.figure(figsize=(18,10))
sns.swarmplot(data = data, x = 'Age_buckets', y = 'MP_count_%')
sns.swarmplot(data = data, x = 'Age_buckets', y = 'Voters_count_%', edgecolor="black",
                linewidth=2,sizes=(40,100))
```
